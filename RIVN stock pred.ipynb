```python
import yfinance as yf
```

```python
rivn = yf.Ticker("RIVN")
```

```python
rivn= rivn.history(period="max")
```

```python
del rivn["Dividends"]
del rivn["Stock Splits"]
```

```python
rivn
```

```python
#will stock go up or down- by tmr price 
rivn["Tomorrow"]= rivn["Close"].shift(-1)
rivn
```

```python
#set target to predict- is tmr price greater than todays price
rivn["Target"]=(rivn["Tomorrow"]>rivn["Close"]).astype(int)
rivn
```

```python
#build model (resistant to overfitting. also pick up non linear relationship)
from sklearn.ensemble import RandomForestClassifier

model= RandomForestClassifier(n_estimators=150, min_samples_split=100,random_state=1)

train= rivn.iloc[:-100]
test= rivn.iloc[-100:]

predictors= ["Close", "Volume","Open","High","Low"]
model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score
import pandas as pd

pred= model.predict(test[predictors])
pred=pd.Series(pred, index=test.index)
pred
```

```python
precision_score(test["Target"], pred)
```

```python
combined=pd.concat([test["Target"], pred], axis=1)
combined.plot()
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=250, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(rivn, model, predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts()/predictions.shape[0]
```

```python
#add more predictors to see if it can improve our model
horizons=[2,5,60,250,500] #in days
new_predictors=[]

for horizon in horizons:
    roll_avg= rivn.rolling(horizon).mean()
    
    ratio_col = f"Close_Ratio_{horizon}"
    rivn[ratio_col]= rivn["Close"]/roll_avg["Close"]
    
    trend_col=f"Trend_{horizon}"
    rivn[trend_col]=rivn.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors+=[ratio_col, trend_col]
```

```python
model=RandomForestClassifier(n_estimators=150, min_samples_split=50,random_state=1)
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:1]
    preds[preds>=.3]=1
    preds[preds<=.3]=0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions= backtest(rivn,model,new_predictors)
```
